# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import random
import numpy as np
import statistics as stats



#Genera un array de 100 elememtos que siga una distribución normal
Data_set  = np.random.normal(size = 100)

#print(Data_set)

#Generar el array con distribución chi-cuadrado

Data_set_chi_2 = np.random.chisquare(df = 3, size = 100)

#print(Data_set_chi_2)

#Principales métricas estadísticas


#Rango
Range_n = max(Data_set)-min(Data_set)
Range_ch = max(Data_set_chi_2)-min(Data_set_chi_2)

print(f"Rango_n: {Range_n}")
print(f"Rango_ch: {Range_ch}")



#Media
mean_n = np.mean(Data_set)
mean_ch = np.mean(Data_set_chi_2)

print("mean_n", mean_n)
print("mean_ch", mean_ch)



#Moda
Moda_n = stats.mode(Data_set)
Moda_ch = stats.mode(Data_set_chi_2)

print(f"Moda_n: {Moda_n}")
print(f"Moda_ch: {Moda_ch}")


#Mediana
Mediana_n = stats.median(Data_set)
Mediana_ch = stats.median(Data_set_chi_2)

print(f"Median_n:{Mediana_n}")
print(f"Median_ch :{Mediana_ch}")


#varianza
Var_n = np.var(Data_set)
Var_ch =np.var(Data_set_chi_2)

print(f"Var_n:{Var_n}")
print(f"Var_ch: {Var_ch}")


#Desviación estándar
std_n = np.std(Data_set)
std_ch = np.std(Data_set_chi_2)

print("std_n", std_n)
print("std_ch", std_ch)

#Medidas de forma

from scipy.stats import skew

skewness_n = skew(Data_set)
skewness_ch = skew(Data_set_chi_2)

print(skewness_n)
print(skewness_ch)


from scipy.stats import kurtosis

kurt_n = kurtosis(Data_set)
kurt_ch = kurtosis(Data_set_chi_2)

print(kurt_n)
print(kurt_ch)













````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO
import numpy as np

import math as mt


data = [4, 2, 5, 8, 6]

#Calcular la media
mean = sum(data)/len(data)

print(f"Media:{mean}")

#Calcular el cuadrado de la distancia a la media de cada dato

distance_2 = []
for x in data:
    distance_2.append((x - mean) ** 2)

print(f"Cuadrado de las distancias a la media: {distance_2}")

distance_2_final = sum(distance_2)

print(distance_2_final)

#Dividir el cuadrado de la diferencia entre el número de valores

valor_casi_final = distance_2_final/len(data)

print(valor_casi_final)

#raiz cuadrada

std_data = mt.sqrt(valor_casi_final)

print(f"STD:{std_data}")
      
std_formula = np.std(data)

print(std_formula)



````
